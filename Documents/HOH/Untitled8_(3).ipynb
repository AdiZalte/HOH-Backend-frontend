```python
import pandas as pd
import numpy as np

# Drop index column if present
if "Unnamed: 0" in ds.columns:
    ds.drop(columns=["Unnamed: 0"], inplace=True)

# -------------------------------
# 1. Handle missing MonthlyIncome
# -------------------------------
ds["MonthlyIncome"].fillna(ds["MonthlyIncome"].median(), inplace=True)

# -------------------------------
# 2. Log transform skewed columns
# -------------------------------
skew_cols = [
    "RevolvingUtilizationOfUnsecuredLines",
    "DebtRatio",
    "MonthlyIncome"
]

for col in skew_cols:
    ds[col] = np.log1p(ds[col])

# -------------------------------
# 3. IQR capping for moderate vars
# -------------------------------
iqr_cols = [
    "NumberOfOpenCreditLinesAndLoans",
    "NumberRealEstateLoansOrLines",
    "NumberOfDependents",
    "age"
]

for col in iqr_cols:
    Q1 = ds[col].quantile(0.25)
    Q3 = ds[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    ds[col] = ds[col].clip(lower, upper)

# -------------------------------
# 4. Domain cap delinquency counts
# -------------------------------
delinq_cols = [
    "NumberOfTime30-59DaysPastDueNotWorse",
    "NumberOfTimes90DaysLate",
    "NumberOfTime60-89DaysPastDueNotWorse"
]

for col in delinq_cols:
    ds[col] = ds[col].clip(upper=20)
```

```python
from sklearn.model_selection import train_test_split

X = ds.drop("SeriousDlqin2yrs", axis=1)
y = ds["SeriousDlqin2yrs"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.1,      # 90:10 split
    random_state=42,
    stratify=y
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print("After SMOTE class distribution:")
print(y_train_sm.value_counts())
```

```python
y_train_sm.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Create scaler
scaler = StandardScaler()

# Fit ONLY on SMOTE training data
X_train_scaled = scaler.fit_transform(X_train_sm)

# Transform test data using same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
# ==============================
# 1. Import Libraries
# ==============================
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

# ==============================
# 2. Compute imbalance weight (optional for XGBoost)
# ==============================
scale_pos_weight = (y_train_sm == 0).sum() / (y_train_sm == 1).sum()

# ==============================
# 3. Train ML Models
# ==============================

# Logistic Regression (needs scaled data)
lr = LogisticRegression(max_iter=1000, class_weight="balanced")
lr.fit(X_train_scaled, y_train_sm)

# Random Forest (no scaling needed)
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)
rf.fit(X_train_sm, y_train_sm)

# KNN (needs scaled data)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train_sm)

# XGBoost (main production model)
xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric="logloss",
    scale_pos_weight=scale_pos_weight
)
xgb.fit(X_train_sm, y_train_sm)

# ==============================
# 4. Evaluation Function
# ==============================
def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall   :", recall_score(y_test, y_pred))
    print("F1 Score :", f1_score(y_test, y_pred))
    print("ROC-AUC  :", roc_auc_score(y_test, y_prob))
    print("-" * 40)

# ==============================
# 5. Evaluate All Models
# ==============================
print("Logistic Regression")
evaluate(lr, X_test_scaled, y_test)

print("Random Forest")
evaluate(rf, X_test, y_test)

print("KNN")
evaluate(knn, X_test_scaled, y_test)

print("XGBoost")
evaluate(xgb, X_test, y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, roc_curve, ConfusionMatrixDisplay
)

# ==============================
# Evaluation Function with Plots
# ==============================
def evaluate_with_plots(model, X_test, y_test, model_name):
    # Predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # ---- Metrics ----
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    print(f"\n===== {model_name} =====")
    print("Accuracy :", acc)
    print("Precision:", prec)
    print("Recall   :", rec)
    print("F1 Score :", f1)
    print("ROC-AUC  :", auc)

    # ---- Confusion Matrix Plot ----
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f"{model_name} - Confusion Matrix")
    plt.show()

    # ---- ROC Curve Plot ----
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{model_name} - ROC Curve")
    plt.legend()
    plt.show()

    return acc, prec, rec, f1, auc
```

```python
results = {}

results["Logistic Regression"] = evaluate_with_plots(lr, X_test_scaled, y_test, "Logistic Regression")
results["Random Forest"] = evaluate_with_plots(rf, X_test, y_test, "Random Forest")
results["KNN"] = evaluate_with_plots(knn, X_test_scaled, y_test, "KNN")
results["XGBoost"] = evaluate_with_plots(xgb, X_test, y_test, "XGBoost")


```

```python
import pandas as pd

results_df = pd.DataFrame(results, index=["Accuracy", "Precision", "Recall", "F1", "ROC-AUC"]).T
print(results_df)

# Optional: bar plot of ROC-AUC comparison
results_df["ROC-AUC"].plot(kind="bar", title="Model ROC-AUC Comparison")
plt.ylabel("ROC-AUC Score")
plt.show()
```
